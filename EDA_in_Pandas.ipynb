```python
df.fillna({"1970 Population" : df["1970 Population"].mean()},inplace=True )
df.fillna({"1980 Population" : df["1980 Population"].mean()},inplace=True )
df.fillna({"1990 Population" : df["1990 Population"].mean()},inplace=True )
df.fillna({"2000 Population" : df["2000 Population"].mean()},inplace=True )
df.fillna({"2010 Population" : df["2010 Population"].mean()},inplace=True )
df.fillna({"2015 Population" : df["2015 Population"].mean()},inplace=True )
df.fillna({"2020 Population" : df["2020 Population"].mean()},inplace=True )
```

```python
df2 = df.groupby('Continent')[['P_1970',
       'P_1980', 'P_1990', 'P_2000',
       'P_2010', 'P_2015', 'P_2020',
       'P_2022']].mean(numeric_only = True).sort_values(by = 'P_2022',ascending = False)
df2
```
